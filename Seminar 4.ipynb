## *Seminar 4: Stochastic differential equations*

## 1. Strong and weak solutions

#### 1.1 Main definitions

Consider a one-dimensional SDE 

$$dX_t = b(t, X_t) dt + \sigma(t, X_t) dW_t, \qquad X_0 = x,$$

that is $$X_t = x + \int_{0}^{t} b(s, X_s) ds + \int_{0}^{t} \sigma(s, X_s) dW_s.$$

 <u> **Definition 1** </u> A **weak solution** of SDE is a pair $(X_t, W_t)$ on some probability space $(\Omega, \mathcal{F}, (\mathcal{F}_t)_{t \geq 0}, P)$, such that 
 
 1. $W_t$ is a Brownian motion with repsect to $\mathcal{F_t}$,
 
 2.  $X_t = x + \int_{0}^{t} b(s, X_s) ds + \int_{0}^{t} \sigma(s, X_s) dW_s$ a.s. for all $t \geq 0$.


 <u> **Definition 2** </u> A **strong solution** of SDE is a weak solution such that $X_t$ is __adapted__ to filtration generated by $W_t$, i.e. $X_t$ is $\mathcal{F_t}^{W}$ measurable.


In other words, 

* for strong solutions we are given a Brownian motion on a given probability space;
* for weak solutions we are free to choose the Brownian motion and the probability space.

 <u> **Definition 3** </u> The SDE has a **weak unique** solution if for any two solutions $(X_t, W_t)$ and $(\tilde X_t, \tilde W_t)$
 the processes $X_t$ and $\tilde X_t$ are equal in distribution, i.e. any finite-dimensional distributions of $X$ and $\tilde X$ coincide, that is 
 for $\forall n > 0$ $\forall t_1 < \dots < t_n$ 

 $$(X_{t_1}, \cdots, X_{t_n}) \stackrel{d}{=} (\tilde X_{t_1}, \cdots, \tilde X_{t_n}).$$
 
 <u> **Definition 4** </u> The SDE has a **strong unique** solution if for any two solutions $(X_t, W_t)$ and $(\tilde X_t, W_t)$, given on the same probability space, $X_t = \tilde X_t$ a.s. for any $t \geq 0$. 


 <u> **Definition 6** </u> SDE is called **homogeneous** if $\sigma, b$ do not depend on $t$, i.e. the SDE looks like 
 $$dX_t = b(X_t) dt + \sigma(X_t) dW_t.$$

#### 1.2 Main results (strong solutions)

<u> **Theorem 1** (Yamada, Watanabe) </u>  

1. __Strong uniqueness__ $\Rightarrow$ __weak uniqueness__.
2. __Strong uniqueness__ and __weak existence__ $\Rightarrow$ __strong existence__.

<u> **Theorem 2** (Cherny, Engelbert) </u>  

__Weak uniqueness__ and __strong existence__ $\Rightarrow$ __strong uniqueness__.

<u> **Theorem 3** (Ito) </u>  

For __strong existence and uniqueness__ it is sufficient to have 

1. $|b(t, x) - b(t, y)| + |\sigma(t, x) - \sigma(t, y)| \leq C|x - y|$, (uniqueness condition)

2. $|b(t, x)| + |\sigma(t, x)| \leq C(1 + |x|)$. (existence condition)

<u> **Example 1** </u>  

$$dX_t = \text{sgn}(X_t)dt + dW_t.$$

Ito's therem is not applicable as $\text{sgn}(x)$ does not satisfy Lipsitz conditions. 

<u> **Theorem 4** (Zvonkin) </u>  

For __strong existence and uniqueness__ it is sufficient to have 

1. $b(t,x)$ is bounded.

2. $\sigma(t, x)$ is continuous. 

3. $| \sigma(t, x) - \sigma(t, y) | \leq C \sqrt{|x-y|}$.

4. $|\sigma(t, x)| \geq \varepsilon > 0$. $\quad$  (Note: inconvenient condition, rarely fulfilled)

<u> **Example 1** (revisited) </u>  

$$dX_t = \text{sgn}(X_t)dt + dW_t.$$

By Zvonkin's theorem there exist strong unique solution. 

<u> **Theorem 5** (Yamada, Watanabe) </u>  

For __strong uniqueness__ it is sufficient to have 

1. $|b(t, x) - b(t, y)| \leq C|x - y|$.

2. $|\sigma(t, x) - \sigma(t, y)| \leq h(|x - y|)$, where $h \uparrow$ and $\int_{0}^{\varepsilon} h^{-2}(x) dx = \infty$ for any $\varepsilon \geq 0$.

#### 1.3 Main results (weak solutions)

<u> **Theorem 6** (Skorokhod) </u>  

For __weak existence__ it is sufficient to have continuous and bounded $b(t, x)$ and $\sigma(t, x)$. 

(Note: inconvenient conditions, boundness of coefficients is rarely fulfilled).

<u> **Theorem 7** (Strookh, Varadhan) </u>  

For __weak existance and uniqueness__ it is sufficient to have 

1. $b(t, x)$ is bounded.

2. $\sigma(t, x)$ is continuous. 

3. $\sigma(t, x) \geq \varepsilon > 0$ for all $t, x$. (Note: inconvenient condition, rarely fulfilled).


<u> __Homogeneous case, zero drift__ </u>

Consider the SDE 

$$dX_t = \sigma(X_t) dW_t, \qquad X_0 = x_0 > 0.$$

Define two sets $I$ and $Z$:

$$I = \{ x \in \mathbb{R}: \int_{x - \varepsilon}^{x + \varepsilon} \dfrac{1}{\sigma^2(y)}dy = \infty, \quad \forall \varepsilon > 0\}, \qquad Z = \{ x \in \mathbb{R}: \sigma(x) = 0\}.$$

<u> **Theorem 8** (Engelbert, Schmidt) </u>  

1. For SDE above we get __weak existence__ if and only if  $I \subseteq Z$.

2. For SDE above we get __weak uniqueness__ if and only if $I = Z$. 

<u> **Theorem 9** (Urusov's martingality test) </u>  

Under the definitions above,

Let 

1. $\sigma(0) = 0$.

2. $\sigma(x) > 0$ for $x > 0$.

3. $\int_{a}^{b} \frac{1}{\sigma^2(x)} < \infty$ for any $0 < a < b.$

4. $X_t$ (which is a weak solution of SDE above, which exists by Engelbert-Schmidt theorem under the assumptions 1-3) is equal to $0$ after reaching the $0$ level. 

Then $X_t$ is a martingale if and only if and only if $\int_{1}^{\infty} x \sigma^{-2}(x) = \infty.$

Some intuition: 

* Why we need to stop the porcess $X_t$ after reaching the $0$ level? Otherwise, if it reflects from $0$, then "on average" it will be greater than zero afther the moment of hitting the zero level, so the martingale condition is violated. 

* How to interpret the divergence of $\int_{1}^{\infty} x \sigma^{-2}(x)$? In fact, if $\sigma(x)$ is large for large $x$, then it means that the process $X_t$ may attain extremely large values, which is possible for local martingales, but not for martingales. 

<u> __Homogeneous case, non-zero drift__ </u>

<u> **Theorem 8** (Engelbert, Schmidt) </u>  

For __weak existence and uniqueness__ of on the interval $(l, r)$, where $- \infty \leq l < r \leq + \infty$, it is sufficient to have 

1. $\sigma(x) \neq 0$ for all $x \in (l, r)$.

2. $\int_{a}^{b} \frac{1 + |b(y)|}{\sigma^2(y)} dy < \infty$ for any $[a, b] \in (l, r)$. 

<u> **Theorem 9** (Feller's test) </u>  

Assume that Engelbert-Schmidt theorem conditions are fulffiled and let 

1. $\tau_{\varepsilon} := \inf \{ t \geq 0: X_t \notin (l + \varepsilon, r - \varepsilon)\}$.

2. $\tau := \lim_{\varepsilon \to 0} \tau_{\varepsilon}$.

3.  $v(x) := \int_{c}^{x} \left( \int_{c}^{y} \exp{ \{-2 \int_{z}^{y} \frac{b(u)}{\sigma^2(u)} du \}} \frac{dz}{\sigma^2(z)} \right)dy$, $\quad x \in (l, r).$

Then $P(\tau = \infty) = 1$ if and only if $v(l+) = v(r-) = \infty.$

In other words, Feller's test gives us necessary and sufficient conditions to check whether the particular process, given by a homogeneous SDE, crosses the particular levels $l$ and $r$. 

#### 1.3 Examples

<u> **Example 2** (Tanaka's, no strong solution) </u>  

$$
\boxed{
    \begin{align*}
    &  {\displaystyle \exists } W \quad {\displaystyle \exists } S \quad !W \quad !S \\
    & + \hspace{0.5cm} - \hspace{0.5cm} + \hspace{0.5cm} -
    \end{align*}
}
$$


 

$$dX_t = \text{sgn}(X_t) dW_t, \quad X_0 = 0,$$

$$\text{sgn}(x) := \begin{cases} +1 , \quad x \geq 0 \\ -1 , \quad x < 0\end{cases}.$$

<u> *Proof*: </u> 


Let $X_t$ be a Brownian motion and $$W_t := \int_{0}^{t} \text{sgn}(X_s) dX_s.$$

Then $W_t$ is a square-integrable martingale, $W_0 = 0$ and by Ito's formula 

$$W_t^2 - t = \int_{0}^{t} 2W_s dW_s + \int_{0}^{t} ds - t = \int_{0}^{t} 2W_s dW_s.$$

Hence $W_t^2 - t$ is a martingale and by Levy's theorem $W_t$ is a Brownian motion. Moreover, 

$$\int_{0}^{t} \text{sgn}(X_s) dW_s = \int_{0}^{t} \text{sgn}^2(X_s) dX_s = X_t.$$

Thus we get that $(X, W)$ is a weak solution of equation $$dX_t = \text{sgn}(X_t) dW_t, \quad X_0 = 0.$$

Notice that $X_t$ is also a Brownian motion, hence we get weak uniqueness as all finitie-dimensional distributions of Brownian motions coincide. 

Now lets show that there is no strong solution. For that, lets show that if $(X, W)$ is a weak solution, then $X_t$ can not be measurable with respect to $\mathcal{F}^{W}_t$. It is enough to show that $W_t$ is measurable with respect to $\mathcal{F}_t^{|X|}$ and $\mathcal{F}_t^{|X|} \subsetneq \mathcal{F}_t^{X}$. 

 <u> **Proposition 1** (Tanaka's formula)</u> 

Let $X_t$ be a Brownian motion. Then 

$$d|X_t| = \text{sgn}(X_t) dX_t + dL_t(X),$$

where $$L_t(X) = \lim_{\varepsilon \to 0}\frac{1}{2 \varepsilon} \int_{0}^{t} \mathbb{I}_{ \{ |X_s| \leq \varepsilon \}} dt$$
is called a local time at $0$ (the local time spent by $X$ at $0$ before time t). 

In integral representation, 

$$|X_t| = \int_{0}^{t} \text{sgn}(X_s) dX_s + L_t.$$
 

<u> *Proof idea:* </u>

Ito's formula con not be applied directly as $| \cdot |$ is not $C^2$. However, $| \cdot |$ can be approximated by 
$$f_{\varepsilon}(x) := \frac{x^2}{2 |\varepsilon|} + \frac{| \varepsilon|}{2}.$$
Then, applying Ito's formula and passing to limit, we get the result. 

<u> *End of proof idea*. </u> 


Now, applying Tanaka's formula, we get that 
$$W_t \equiv \int_{0}^{t} \text{sgn}(X_s) dX_s = |X_t| - L_t(X)$$

and both $|X_t|$ and  $L_t(X)$ are measurable with respect to $\mathcal{F}_t^{|X|}$. Hence, $W_t$ is measurable with respect to $\mathcal{F}_t^{|X|}$. 

At the same time, $\mathcal{F}_t^{|X|} \subsetneq \mathcal{F}_t^{X}$ as, for instance, $\{ X_1 < 0\} \notin \mathcal{F}_1^{|X|}$. Therefore, $X_t$ is not measurable wrt to $\mathcal{F}_t^{W}$, hence there is no strong solution. 

Applying Yamada-Watanabe theorem, we obtain that there is no strong uniqueness. It completes the proof.

<u> *End of proof*. </u> 

<u> **Example 3** (no strong and weak uniqueness) </u>  

$$
\boxed{
    \begin{align*}
    & {\displaystyle \exists } W \quad {\displaystyle \exists } S \quad !W \quad !S \\
    & + \hspace{0.5cm} + \hspace{0.5cm} - \hspace{0.5cm} -
    \end{align*}
}
$$

$$dX_t = \mathbb{I}_{ \{ X_t \neq 0 \}} dW_t, \quad X_0 = 0.$$

<u> *Proof*: </u> 

Two strong solutions $(0, W)$ and $(W, W)$. Indeed, 

$$(0, W): \qquad X_t = \int_{0}^{t} \mathbb{I}_{\{ X_s \neq 0 \}} dW_s = \int_{0}^{t} 0 dW_s = 0.$$
$$(W, W): \qquad X_t = \int_{0}^{t} \mathbb{I}_{\{ X_s \neq 0 \}} dW_s = \int_{0}^{t} 1 dW_s = W_t.$$

<u> *End of proof*. </u> 

<u> **Example 4** (Girsanov's, no strong and weak uniqueness) </u>  

$$
\boxed{
    \begin{align*}
    & {\displaystyle \exists } W \quad {\displaystyle \exists } S \quad !W \quad !S \\
    & + \hspace{0.5cm} + \hspace{0.5cm} - \hspace{0.5cm} -
    \end{align*}
}
$$

$$dX_t = |X_t|^{\alpha} dW_t, \quad X_0 = 0, \quad \alpha < 1/2$$

<u> *Proof*: </u> 

Let $\widetilde W_t$ be a Brownian motion and 
$$A_t := \int_{0}^{t} |\widetilde W_s|^{-2\alpha} ds, \quad T_t = \inf \{ s  \geq 0: A_s > t\}, \quad X_t := \widetilde W_{T_t.}$$

It can be proved that the process $T_t$ is a continuous and strictly increasing stopping time, and $A_t \to \infty$ as $t \to \infty$. 

$X_t$ is a continuous local martingale with respect to to $\mathcal{F}^{W}_{T_t}$ and $\langle  X \rangle _t  = T_t$. 

Hence, $B_t := \int_{0}^{t} |X_s|^{-\alpha} dX_s$ is also a local martingale and, by Ito's formula, 


$$B^2_t - t = \int_{0}^{t} 2B_sdB_s + \int_{0}^{t} |X_s|^{-2\alpha} dT_s - t = \int_{0}^{t} 2B_s |X_s|^{-\alpha} dX_s + \int_{0}^{t} |\tilde W_{T_s}|^{-2\alpha} dT_s - t = \\ \\

=  \int_{0}^{t} 2B_s |X_s|^{-\alpha} dX_s + \int_{0}^{T^{-1}_t} |\tilde W_{s}|^{-2\alpha} ds - t =  \int_{0}^{t} 2B_s |X_s|^{-\alpha} dX_s + t - t =  \int_{0}^{t} 2B_s |X_s|^{-\alpha} dX_s.$$

Hence, $B^2_t - t$ is a local martingale and by Levy's characterization theorem $B_t$ is a Brownian motion and hence $(X, B)$ is a weak solution.

Another strong and weak solution is $(0, B)$. 

<u> *End of proof*. </u> 

## 2. Applications to important examples 

#### 2.1 Black Scholes

$$dX_t = \sigma X_t dB_t, \quad X_0 = x_0 > 0.$$

__Strong and weak existence and uniqueness__

* Applying Ito's theorem with $\sigma(x) = \sigma x$, which is, obviously, of linear growth and Lipsitz, we get __strong existence and uniqueness__.
* __Weak existence and uniqueness__ automatically follows from __strong existence and uniqueness__.


__Positiveness__

* Apply Feller's test with $(l, r) = (0, \infty)$:

$$v(x) := \int_{c}^{x} \left( \int_{y}^{c} \frac{dz}{\sigma^2 z^{2}} \right)dy \propto \int_{c}^{x} \left (\frac{1}{y} - \frac{1}{c} \right) dy \propto \ln(x) + \tilde c$$


* $v(0) = \infty$, hence the process does not reach zero and always remains  positive. 

__Martingality__

* Appling Urusov's test, we get the martingality




$$\int_{1}^{\infty} (\sigma^2 z)^{-1} = \infty,$$


*Note: Obviously, for GBM from Black Scholes model we do not need such strong results as Feller's and Urusov's tests. Positiveness and martingality immediately follow from the explcit solution:*

$$X_t = X_0 \exp{ \{-\sigma^2/2 t + \sigma W_t \}}$$

#### 2.2 CEV

$$dX_t = |X_t|^{\alpha} dB_t, \quad X_0 = x_0.$$

__Strong and weak existence and uniqueness__

* From Engelbert-Schmidt theorem with $\sigma(x) = |x|^{\alpha}$, we get that $I \subseteq Z$ and we have __weak existence__. 

* For $\beta \geq 1/2$, the __strong uniqueness__ follows from Yamada-Watanabe theorem with $h(x) = c\sqrt{x}$. Hence, applying another Yamada-Watanabe theorem, we get __strong existence__. Thus, we get unique solution. 

* For $\beta < 1/2$ with $x_0 = 0$, we have __strong and weak existence__, but no uniqueness. 

__Positiveness__


* Apply Feller's test with $(l, r) = (0, \infty)$:

$$v(x) := \int_{c}^{x} \left( \int_{y}^{c} \frac{dz}{z^{2 \alpha}} \right)dy = \int_{c}^{x} \begin{cases} \ln c - \ln y, \quad \alpha = 1/2 \\ c^{-2\alpha + 1} - y^{- 2\alpha + 1}, \quad \alpha \neq 1/2\end{cases} dy.$$

* We see that $v(0) = \infty$ if and only if $\alpha \geq 1$. Hence, if $\alpha \geq 1$, the solution does not reach zero and reachs zero if and only if $\alpha < 1$. 

__Martingality__


* Appling Urusov's test, 

$$\int_{1}^{\infty} z^{1 - 2\alpha} = \infty \iff \alpha \leq 1,$$

* we get the martingality of $X_t$ iff $\alpha \leq 1$ and only a local martingality iff $\alpha > 1$.

#### 2.3 CIR

$$dX_t = \kappa (\theta - X_t) dt + \sigma \sqrt{X_t^{+}} dW_t, \quad X_0 = x_0 > 0.$$


where parameters $\kappa > 0, \theta > 0, \sigma > 0$.

__Strong and weak existence and uniqueness__


* From Engelbert-Schmidt theorem with $\sigma(x) = \sigma \sqrt{x}$ and $b(x) = \kappa (\theta - x)$ we have __weak existence and uniqueness__ as for for any $[a, b] \in (0, \infty)$

$$\int_{a}^{b} \dfrac{1 + |\kappa (\theta - x)|}{\sigma^2 x} dy < \infty.$$ 



__Positiveness__

* Apply Feller's test with $(l, r) = (0, \infty)$:

$$v(x) := \int_{c}^{x} \left( \int_{c}^{y} \exp{ \{-2 \int_{z}^{y} \frac{\kappa (\theta - u)}{\sigma^2 u} du \}} \frac{dz}{\sigma^2 z} \right)dy \simeq \int_{c}^{x} (y^{-2 \kappa \theta / \sigma^2} - 1).$$

* $v(0) = \infty \iff 2 \kappa \theta \geq \sigma^2$, hence the process $X_t$ reaches zero if and only if $2 \kappa \theta < \sigma^2 $ and otherwise remain strictly positive. 

* $^{*}$ Using analysis of singular points (beyond this course), it can be shown that if $2 \kappa \theta < \sigma^2 $, then the solution can be continued for all $t \geq 0$ and "reflects" at zero. 

__Martingality__

* Obviously, there is no martingality because of drift term.
